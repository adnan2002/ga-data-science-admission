```python
import pandas as pd

df = pd.read_csv("./data/DSI_kickstarterscrape_dataset.csv", encoding="latin-1")

print("Number of rows", len(df))
df.head()
```

```python
# Answering Question 1
# What is the mean (total) pledge that projects get? (not per backer)
# Your answer may vary by +/- 5%
```

```python
# First Step: Let's clean the data of the 'pledged' column and remove any non-numeric values

pledged_df = df.copy()

# Remove any non-numeric values
pledged_df['pledged'] = pledged_df['pledged'].apply(lambda x: x if isinstance(x, (int, float)) else None)

# Drop rows where 'pledged' is None/NaN
pledged_df = pledged_df.dropna(subset=['pledged'])

print("Number of rows after removing non-numeric values: ", len(pledged_df))
```

```python
# Seconds Step: Check if all project ids are unique

print("Are all project ids unique? ", pledged_df["project id"].nunique() == len(pledged_df["project id"]))
```

```python
print("Number of duplicated rows: ", pledged_df["project id"].duplicated().sum())
```

```python
pledged_df[pledged_df["project id"].duplicated(keep=False)].sort_values("project id")
```

```python
# Step 3: Dropping duplicated rows
pledged_df = pledged_df.drop_duplicates(subset=["project id"])

print("Number of rows after dropping duplicated rows: ", len(pledged_df))
```

```python
# Check again if all project ids are unique
print("Are all project ids unique? ", pledged_df["project id"].nunique() == len(pledged_df["project id"]))
print("Number of duplicated rows: ", pledged_df["project id"].duplicated().sum())
pledged_df[pledged_df["project id"].duplicated(keep=False)]
```

```python
# Step 4: Calculate the mean pledge

mean_pledged = pledged_df['pledged'].mean()

print("Mean pledge: ", mean_pledged)

```
